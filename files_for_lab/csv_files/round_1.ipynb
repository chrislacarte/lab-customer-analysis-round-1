# Read the three files into python as dataframes

```python
import pandas as pd

file1 = pd.read_csv('./file1.csv')
file2 = pd.read_csv('./file2.csv')
file3 = pd.read_csv('./file3.csv')


```

# Show the DataFrame's shape.

```python
print(file1.shape)
print(file2.shape)
print(file3.shape)
```

# Standardize header names. and Rearrange the columns in the dataframe as needed

```python
lst1 = []
for col in file1.columns:
    lst1.append(col.lower())
file1.columns = lst1 

lst2 = []
for col in file2.columns:
    lst2.append(col.lower())
file2.columns = lst2

lst3 = []
for col in file3.columns:
    lst3.append(col.lower())
file3.columns = lst3


#change column names just for 'state' in file3
file3 = file3.rename(columns={'state':'st'})


#Re-order the columns like file1

file2_ord = file2[['customer', 'st', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]
file3_ord = file3[['customer', 'st', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]
print(file1.columns)
print(file2_ord.columns)
print(file3_ord.columns)
```

# Concatenate the three dataframes

# Eliminate "%" from column so it doesn't create more NaN values  and transform them to numeric values as the% doiesnt reciognise as a number, but a string

```python
file1['customer lifetime value'] = file1['customer lifetime value'].str.replace('%', '')
file2['customer lifetime value'] = file2['customer lifetime value'].str.replace('%', '')
file1['customer lifetime value'] = pd.to_numeric(file1['customer lifetime value'], errors='coerce')
file2['customer lifetime value'] = pd.to_numeric(file2['customer lifetime value'], errors='coerce')
file3['customer lifetime value'] = pd.to_numeric(file3['customer lifetime value'], errors='coerce')

file1['customer lifetime value'] = file1['customer lifetime value']/100

file2['customer lifetime value'] = file2['customer lifetime value']/100
```

```python
df = pd.concat([file1,file2,file3], axis=0)
# We reset the index to have it as a reference in case we need it 
df = df.reset_index(drop=True)
print(df.shape)
df
```

# Perform the data cleaning operations mentioned so far in class

# Delete the column education and the number of open complaints from the dataframe

```python
df = df.drop(['education','number of open complaints','customer'], axis=1)
print(df.head())
df.shape
```

# Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change `dtype` to `numerical` type.

```python
# df['customer lifetime value'] =  float(df['customer lifetime value'].strip('%'))/100 
df['customer lifetime value'] = df['customer lifetime value']*100
df['customer lifetime value'] =  pd.to_numeric(df['customer lifetime value'], errors='coerce')

df._get_numeric_data()
```
